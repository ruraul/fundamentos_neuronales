```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(a):
  return 1/(1+np.exp(-a))
```

```python
def steep(x, derivate = False):    
    return np.piecewise(x,[x<0.0,x>=0.0],[0,1])
```

```python
def relu(x):
  return np.maximum(0,x)
```

```python
def tanh(x):
  return np.tanh(x)
```

```python
x = np.linspace(10,-10,100)
```

```python
plt.plot(x,sigmoid(x))
```

```python
plt.plot(x,steep(x))
```

```python
plt.plot(x, relu(x))
```

```python
plt.plot(x, tanh(x))
```
